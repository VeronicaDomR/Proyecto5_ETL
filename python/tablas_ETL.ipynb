```python
import pandas as pd

# Cargar las tablas de dimensiones
df_producto = pd.read_csv('/content/dim_product.csv')
df_customer = pd.read_csv('/content/dim_customer.csv')
df_competidores = pd.read_csv('/content/dim_competitors.csv')
df_supermarket_link = pd.read_csv('/content/supermarket_links.csv')
df_envio = pd.read_csv('/content/dim_envio.csv')
df_tiempo = pd.read_csv('/content/dim_tiempo.csv')

# Asegúrate de que cada DataFrame tenga una columna 'row_num' para alinear
df_producto['row_num'] = df_producto.index + 1
df_customer['row_num'] = df_customer.index + 1
df_competidores['row_num'] = df_competidores.index + 1
df_supermarket_link['row_num'] = df_supermarket_link.index + 1
df_envio['row_num'] = df_envio.index + 1
df_tiempo['row_num'] = df_tiempo.index + 1

# Selecciona solo las columnas relevantes y asegúrate de que todas tengan 'row_num'
df_producto = df_producto[['product_id', 'row_num']]
df_customer = df_customer[['customer_id', 'row_num']]
df_competidores = df_competidores[['company_id', 'row_num']]
df_supermarket_link = df_supermarket_link[['market_id', 'row_num']]
df_envio = df_envio[['envio_id', 'row_num']]
df_tiempo = df_tiempo[['order_id', 'row_num']]

# Realizar el merge para combinar todas las IDs en un único DataFrame alineado
df_combined = df_producto.merge(df_customer, on='row_num', how='left')
df_combined = df_combined.merge(df_competidores, on='row_num', how='left')
df_combined = df_combined.merge(df_supermarket_link, on='row_num', how='left')
df_combined = df_combined.merge(df_envio, on='row_num', how='left')
df_combined = df_combined.merge(df_tiempo, on='row_num', how='left')

# Ordenar las columnas finales
df_combined = df_combined[['product_id', 'customer_id', 'company_id', 'market_id', 'envio_id', 'order_id', 'row_num']]

# Guardar el DataFrame resultante
df_combined.to_csv('/content/tabla_combined.csv', index=False)

# Ver el DataFrame final
display(df_combined)
```
